```python
x_all = newData.drop(['FTR'],1)
y_all = newData['FTR']
```

```python
from sklearn.preprocessing import scale #used to standardize data
columns = [['HTP','ATP','HTGD','ATGD','DiffFormPts']]
for column in columns:
    x_all[column] = scale(x_all[column])
```

```python
setIndex = pd.DataFrame(index = x_all.index)
```

```python
for col, colData in x_all.iteritems():

    if colData.dtype == object:
            colData = pd.get_dummies(colData, prefix = col) 
            
    setIndex = setIndex.join(colData)

x_all = setIndex
```

```python
print ("Number of features: "+str(len(x_all.columns)))
print (list(setIndex.columns))
```

```python
x_all
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_all, y_all, test_size = 600,stratify = y_all)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.metrics import f1_score 

def lPredicts(algo, features, target):
   # Makes predictions based on accuracy metrics and the F1 score.
    y_predict = algo.predict(features)

    return f1_score(target, y_predict, pos_label='H'), sum(target == y_predict) / float(len(y_predict))

def train_predicts(algo, x_all, y_all, x_test, y_test):
    # Train the models
    algo.fit(x_train, y_train)
    
   #Outputs metrics for training dataset
    f1, accuracy = lPredicts(algo, x_train, y_train)
    print ("The F1 score for the training dataset is "+str(f1))
    print ("The accuracy score for the training dataset is "+str(accuracy))
   #Outputs metrics for testing dataset
    f1, accuracy = lPredicts(algo, x_test, y_test)
    print ("The F1 score for the testing dataset is "+str(f1))
    print ("The accuracy score for the testing dataset is "+str(accuracy))
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

LR = LogisticRegression(random_state = 10) 
RF = SVC(random_state = 200, kernel='rbf')

train_predicts(LR, x_train, y_train, x_test, y_test)
print ('')
train_predicts(RF, x_train, y_train, x_test, y_test)
print ('')

```
